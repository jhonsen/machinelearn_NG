### The Gradient Descent method  
To take into account the intercept term \theta0, we add an additional first column to X and set it to all ones. This allows us to treat \theta0 as simply another feature.


```python
# Add one's as feature1
feat1 = np.ones(m)  # m is len(y)

X = np.column_stack((feat1, X))
```

```python
# define the cost function

def computeCost(X,y, theta = [[0], [0]]):
    m = y.size
    h = X.dot(theta)    
    J = (1 / (2 * m)) * np.sum( np.square(h - y))  
    return J
```

```python
# Try compute J function

# Start theta as 0,0 in a 2x1 matrix 
theta = np.zeros((2, 1))   

print('Testing the cost function ...\n')
J = computeCost(X, y, theta)
print('With theta = [[0], [0]]\n Cost computed = %f\n' %J)
print('Expected Cost value (approx)  = 32.07')
```

```python
# Further testing the function, with theta -1,2 in a 2x1 matrix

print('Testing the cost function ...\n')
J = computeCost(X, y, theta = [[-1],[2]])
print('With theta = [[-1], [2]]\n Cost computed = %f\n' %J)
print('Expected cost value (approx) = 54.24\n')
```

```python
# Setting up gradient Descent parameters

iterations = 1500
alpha = 0.01
```

```python
def gradientDescent(X, y, theta, alpha = 0.01, iter = 1500):
    ''' gradient descent
    theta: initial theta
    alph: learning rates, default 0.01
    iter : iterations, default 1500
    
    '''
    m = y.size
    Jhist = np.zeros(iter)
    
    for i in np.arange(iter):
        h = X.dot(theta)
        theta = theta - alpha*(1/m) * (X.T.dot(h-y))
        Jhist[i] = computeCost(X, y, theta)
    return theta, Jhist
                
```

```python
# Running Gradient Descent
# and Compare resulting theta with expected values

print('Running Gradient Descent,\n with alpha = %s and %s iterations\n' %(alpha,iterations))
res_theta, res_J = gradientDescent(X,y,theta, alpha, iterations)

print('Theta found by gradient descent:\n%s\n' %res_theta)
print('Expected theta values (approx)\n -3.6303\n 1.1664\n')
```

Plot the result of gradient descent on the scatterplot of data
  

```python
fig,ax = plt.subplots(figsize = (8,5))
ax.plot(X[:,1], y, 'r*', label ='dataset' )
ax.plot(X[:,1], X.dot(res_theta), 'b-', label = 'linear regression')
ax.legend()
ax.set_xlabel('Population of City in 10,000s ')
ax.set_ylabel('Profit in 10,000s')
```

Predicting values for population sizes of 35,000 and 70,000

```python
predict1 = (np.array([1, 3.5]) *10000).dot(res_theta)
predict2 = (np.array([1, 7.0]) *10000).dot(res_theta)
print('For population 35,000, we predict %.2f' %predict1)
print('For population 70,000, we predict %.2f' %predict2)
```

### Feature Normalization 

```python
def featureNormalize(X):
    xNorm = X
    mu = np.zeros((1, np.size(X, 1)))
    sigma = np.zeros((1, np.size(X, 1)))
    
    for col in np.arange(np.size(X, 1)):
        mu[:, col] = np.mean(X[:,col])
        sigma[:, col] = np.std(X[:,col])
        xNorm[:,col] = (X[:,col] - mu[0,col]) / sigma[0,col]
    
    return (xNorm, mu, sigma)
```

```python
## Loading data 

data = np.loadtxt('./ex1data2.txt',delimiter = ',')
```

```python
X = np.stack(data[:, 0:2])
y = np.row_stack(data[:, 2])

X.shape, y.shape
```

```python
## Print out some data points

print('First 10 examples from the dataset: \n')
print(X[:10,:]) 
```

```python
## Scale features and set them to zero mean
print('Normalizing Features ...\n')

Xnorm, mu, sigma = featureNormalize(X)

# Add one's as feature1
feat1 = np.ones(len(y))  

## Add intercept term to X
X = np.column_stack((feat1, Xnorm)) 
                                        # Carefeul !!! run this cell only once
X[:5,:]
```

```python
## Running Gradient Descent
print('Running gradient descent...\n')


# Parameters for Gradient descent
alpha = 1.0
iterations = 100

# Init Theta and Run Gradient Descent 
theta = np.zeros((3, 1))
theta, J_history = gradientDescent(X, y, theta, alpha, iterations)

# Then, Plot the convergence graph
fig, ax = plt.subplots()
ax.plot(np.arange(1,iterations+1), J_history, '-b')
ax.set_xlabel('Number of iterations')
ax.set_ylabel('Cost J')

# Display gradient descent's result
print('Theta computed from gradient descent: \n')
print(theta)
print('\n')
```

```python
# Estimate the price of a 1650 sq-ft, 3 br house

# the first column of X is all-ones. Thus, it doesnot need to be normalized
# 

# normalizing [1650 3]
est = np.array([[1650, 3]])
norm_est = np.divide(np.subtract(est,mu),sigma)
    
```

```python
   
## Include 1's to first column
input_est = np.column_stack((np.ones(1), norm_est[:,0], norm_est[:,1]))

price = input_est.dot(theta)

print('Predicted price of a 1650 sq-ft, 3 br house (using gradient descent) is\n $%f\n' %price)

```

### The normal equation method
- try this method to solve for theta

```python
## Load dataset

data = np.loadtxt('ex1data2.txt', delimiter=',')
Xtemp = np.row_stack((data[:,0:2]))
y = np.row_stack((data[:,2]))
```

```python
# Add intercept term to X

X = np.column_stack((np.ones(47), Xtemp))
```

```python
## Normal equation function

def normalEqn(X,y):
    theta = np.linalg.pinv((X.T.dot(X))).dot(X.T).dot(y)    
    return theta

```

```python
## Calculate the parameters from the normal equation

theta = normalEqn(X, y)

## Display normal equation's result

print('Theta computed from the normal equations: \n')
print(theta)
print('\n')
```

```python
# Finally, let's estimate the price of 1650 sq-ft, 3br house

est = np.array([[1, 1650, 3]])
price = est.dot(theta)

print('Predicted price for a 1650 sq-ft and 3 br house is\n  $%.2f' %price)
```

Basically, using this method we arrive at house __$293k__, which is about the same as the gradient descent method used earlier
